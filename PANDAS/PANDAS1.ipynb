```python
import pandas as pd
data={ "NAME":["TYAGI","UDAY","ANSH","ARJUN"],"SALARY":[100000,20000,700000,90000],"AGE":[19,20,34,56],"CITY":["GHAZIABAD","DELHI","LUCKNOW","BANGLORE"],"OCCUPATION":["ENGINEER","CRICKETER","DOCTOR","NITHSLLA"]}
df=pd.DataFrame(data)
doublecon=df[(df['AGE']>25) & (df['SALARY']>20000)]
print("EMPLOYEE SALARY GREATER THAN 20000 nd age greater than 25")
print(doublecon)
```
