```python
#group by
import pandas as pd
data={"NAME":["UDAY","TYAGI","SAKHI","MAHI","VIRAT"],"ROLL NO":[1,24,5,45,4],"COMPANY":["GOOGLE","AMAZON","TCS","SEVEN","PUMA"],"SALARY":[5000000,10000000,500000,10000000,10000000]}
df=pd.DataFrame(data)
grouped=df.groupby("ROLL NO")["SALARY"].sum()
print(grouped)
```

```python
#group by
import pandas as pd
data={"NAME":["UDAY","TYAGI","SAKHI","MAHI","VIRAT"],"ROLL NO":[1,24,4,45,4],"COMPANY":["GOOGLE","AMAZON","TCS","SEVEN","PUMA"],"SALARY":[5000000,10000000,500000,10000000,10000000]}
df=pd.DataFrame(data)
grouped=df.groupby(["ROLL NO","NAME"])["SALARY"].sum()
print(grouped)
```

```python
#merging & joining(inneer)
import pandas as pd
df_customers=pd.DataFrame({
    "CUSTOMER_ID":[1,2,3],
    "NAME":["SURESH","KALPESH","DINESH"]
        })
df_orderdata=pd.DataFrame({
    "CUSTOMER_ID":[1,2,4],
    "order_amount":[2300,1299,4600]
        })
df_merged=pd.merge(df_customers,df_orderdata,on="CUSTOMER_ID",how="inner")
print(df_merged)
```

```python
#outer join
import pandas as pd
df_customers=pd.DataFrame({
    "CUSTOMER_ID":[1,2,3],
    "NAME":["SURESH","KALPESH","DINESH"]
        })
df_orderdata=pd.DataFrame({
    "CUSTOMER_ID":[1,2,4],
    "order_amount":[2300,1299,4600]
        })
df_merged=pd.merge(df_customers,df_orderdata,on="CUSTOMER_ID",how="outer")
print(df_merged)
```

```python
#right & left join
import pandas as pd
df_customers=pd.DataFrame({
    "CUSTOMER_ID":[1,2,3],
    "NAME":["SURESH","KALPESH","DINESH"]
        })
df_orderdata=pd.DataFrame({
    "CUSTOMER_ID":[1,2,4],
    "order_amount":[2300,1299,4600]
        })
df_merged=pd.merge(df_customers,df_orderdata,on="CUSTOMER_ID",how="right")
print(df_merged)
```

```python
#CONCATENATION
import pandas as pd
df_reg1=pd.DataFrame({
    "customer_id":[1,2],
    "Names":["UDAY TYAGI","RAJU"]
        })
df_reg2=pd.DataFrame({
    "customer_id":[3,4],
    "Names":["shyam","babu rao"]
        })
df_concat=pd.concat([df_reg1,df_reg2],axis=1,ignore_index=True)
print(df_concat)
    
```
