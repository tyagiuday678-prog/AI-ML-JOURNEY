```python
#missing values handlation
import pandas as pd
data={ "NAME":["TYAGI",None,"ANSH","ARJUN"],"SALARY":[100000,None,700000,90000],"AGE":[None,20,34,56],"CITY":["GHAZIABAD","DELHI","LUCKNOW","BANGLORE"],"OCCUPATION":["ENGINEER","CRICKETER","DOCTOR","NITHSLLA"]}
df=pd.DataFrame(data)
print(df.isnull())
print("count of missing values in each column-:")
print(df.isnull().sum())
```

```python
import pandas as pd
data={ "NAME":["TYAGI",None,"ANSH","ARJUN"],"SALARY":[100000,None,700000,90000],"AGE":[None,27,34,56],"CITY":["GHAZIABAD","DELHI","LUCKNOW","BANGLORE"],"OCCUPATION":["ENGINEER","CRICKETER","DOCTOR","NITHSLLA"]}
df=pd.DataFrame(data)
df.fillna({"AGE":df["AGE"].mean()},inplace=True)
df.fillna({"SALARY": df["SALARY"].mean()}, inplace=True)
print(df)
```

```python
#interpolation
import pandas as pd
data={ "NAME":["TYAGI",None,"ANSH","ARJUN"],"SALARY":[100000,None,700000,90000],"AGE":[None,27,34,56],"CITY":["GHAZIABAD","DELHI","LUCKNOW","BANGLORE"],"OCCUPATION":["ENGINEER","CRICKETER","DOCTOR","NITHSLLA"]}
df=pd.DataFrame(data)
#linear,polynomial,time
df['SALARY']=df['SALARY'].interpolate(method="linear")
print(df)
```
