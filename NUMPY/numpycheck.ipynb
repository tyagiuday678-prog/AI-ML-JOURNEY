```python
import numpy as  np
arr=np.array([1,2,4,np.nan,np.nan,6])
new_arr=np.nan_to_num(arr,nan=10)
print(new_arr)
print(np.isnan(arr))
print(np.nan!=np.nan)
```

```python
import numpy as np
arr=np.array([1,2,3,4,5,65,9,-np.inf,np.inf])
print(np.isinf(arr))
cleaned_arr=np.nan_to_num(arr,posinf=100,neginf=-100)
print(cleaned_arr)
```
